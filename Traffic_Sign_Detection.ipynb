**Convolutional Neural Network**

Feature extraction and classification have long been major study areas in the science of computer vision. In traditional image processing tasks, pre-designed features based on statistical regularities or previous knowledge are frequently used as extracted features. Thus, it is unable to precisely and completely portray the information contained in the original image. The gradient descent approach may be used to train parameters in CNN's end-to-end learning model. A well-trained CNN may get a deeper understanding of the image's characteristics, making it a more effective black box for feature extraction.
As a crucial subset of neural networks, CNN expands the idea of receptive field and shared weights, drastically reducing the training parameters while simultaneously simplifying the network topology. By distributing the weight of the convolutional kernel, the features of each layer are created from the local region (receptive field) of the preceding layer. Because of these traits, the CNN is more suited than other neural networks to learn and represent visual information. It can also maintain translation and scale invariance to a certain degree.

Convolutional neural networks perform better on huge datasets, which leads to the progressive abolition of older approaches based on SVM and the resolution of recognition issues in certain circumstances

1. Previous Implementation

Using the gathered dataset of Chinese traffic signs and the regional depth convolutional neural network (CNN), Qian et al. trained the traffic sign data and were successful in achieving a high accuracy recognition rate (Rongqiang Qian et al., 2015).



2. Advantages

The multilevel representations from pixel to high-level semantic features learned by a hierarchical multistage structure, known as a hierarchical feature representation  can be automatically learned from data and hidden factors of input data can be disentangled through multilevel nonlinear mappings.

A deeper architecture offers considerably more expressive capabilities than conventional shallow models.

The CNN architecture offers the chance to jointly optimise a number of linked activities (e.g., fast R-CNN combines classification and bounding box regression into a multitask learning manner).

Some traditional computer vision issues may be reformulated as high-dimensional data transform problems and tackled from a different angle by taking advantage of the deep CNNs' substantial learning potential (Zhao et al., 2019).



3. Disadvantages

Due to the FC layers, the CNN needs an input picture of a constant size (for example, 227227), which necessitates recalculating the whole CNN for each assessed area, which takes a long time during the testing period.

R-CNN training involves a multi-stage pipeline. An object proposal convolutional network (ConvNet) is first polished. Then, SVMs are used to integrate with ConvNet features in place of the softmax classifier that was trained through fine-tuning. The final step is to train bounding-box regressors.

Training is costly both in terms of money and time. Features are taken out of several area suggestions and kept on the disc. A relatively modest training set will take a long time to process with highly deep networks like VGG16. The storage memory needed by these features should also be a concern at the same time.

Despite the fact that selective search might produce region ideas with relatively high recalls, the acquired region proposals still require time to process (around 2 s to extract 2000 region proposals) (Zhao et al., 2019).

4. Input data required and expected output

Input data for a CNN in a traffic sign recognition system is typically image data. The images used for training and testing the CNN should be labeled with the appropriate traffic sign class. The dataset should have large number of images for each traffic sign class to avoid overfitting, and those images should be pre-processed in way that makes them suitable for training and testing of the network. It may include resizing images to a standard size, normalizing the pixel values, and augmenting the data set by applying random transformations to the images such as rotation, scaling and flipping.

The expected output from a CNN in a traffic sign recognition system is a probability for each class. The class with the highest probability is the predicted class for a given image.

Evaluating the CNN model with multiple metrics on dataset like accuracy, recall, precision, F1-score, log loss, which are generated using the confusion matrix. A common way to evaluate the performance of a CNN in a traffic sign recognition system is to measure the accuracy of the network on a held-out test set.

Also, the performance of a CNN in traffic sign recognition may be affected by factors such as the quality of the input data and the size and architecture of the network.

**Support Vector Machine**


Since its emergence in the 1990s, SVM, one of the most effective machine learning algorithms, has been mostly employed for pattern identification. SVM has evolved into a very useful method as a result of its extensive range of applications, including face and speech recognition, face identification, and picture recognition.

Pattern recognition is frequently used to categorise data, depending on either a priori information or statistical data extracted from raw data collection, making it an exceptionally useful tool for data separation in many domains. It is a supervised form of machine learning method that creates a model based on a set of highlighted instances designated as belonging to a category out of several pre-defined categories that predicts the category of a new example.

The support vector machine (SVM) primarily focuses on categorising patterns, which implies that this technique is utilised to categorise various patterns . There are now several types of patterns, including linear and non-linear. Non-linear patterns must be further modified in order to be easily differentiated from linear patterns since they cannot be easily distinguished or can only be separated in high dimensions (Karamizadeh et al., 2014).

The Maldonado-Bascón group from Spain had put a lot of effort into creating a traffic sign recognition system for maintaining road signs. They introduced a support vector machine-based system for automatic road sign identification and recognition in 2007. Distance to Borders feature vectors were supplied into linear SVMs for the form classification and colour segmentation based on thresholding in HSV space, while Gaussian-Kernel SVMs was employed for the classification of pictograms. Then, in, they put forth a new shape feature based on the absolute value of the FFT of the normalised signature of blobs, which shown outstanding resistance to scaling, rotation, occlusion, projective deformation, and noise (Fu & Huang, 2010).

2. Advantages

Support vector machines perform similarly well when there is a discernible margin of class dissociation.

High dimensional rooms are more productive.

When the quantity of dimensions exceeds the quantity of examples, it works well.

Similar to memory systematic is support vector technology (2020).

The SVM algorithm can provide a singular result when the optimality is rounded. This is a key distinction between SVM and neural networks, which generate several solutions based on local minima and are hence unreliable across various samples.

In this procedure, making the data linearly separable doesn't need any assumptions about the actual type of transformation. This is caused by the kernel's presence of a non-linear transformation. Since this transition is based on a solid theoretical foundation, further human knowledge is not required (Karamizadeh et al., 2014).

3. Disadvantages

For huge data sets, the support vector machine approach is unacceptable.

When the target classes overlap and the data set includes more sound, it does not operate very well.

4. Input data required and expected output

The input data required for SVMs are a set of labeled feature vectors which can be extracted from image data by using techniques such as texture, shape and color. The expected output is a classification into one of the predefined traffic sign classes. SVMs are more computationally expensive than other algorithms such as KNN. Performance of the SVM classifier highly depends on the choice of kernel function, type and number of features used, and the parameters used to train the SVM. The model's performance can be evaluated by using metrics such as accuracy, precision, recall and F1-score, and visualized by creating ROC curve. In summary, SVMs can be effective in traffic sign recognition but the performance highly depends on the selection of features, kernel function and parameters, and can be computationally expensive.

**K Nearest Neighbors**


The k-nearest neighbours algorithm, sometimes referred to as KNN or k-NN, is a supervised learning classifier that employs proximity to produce classifications or predictions about the grouping of a single data point. Although it may be applied to classification or regression issues, it is commonly employed as a classification method since it relies on the idea that comparable points can be discovered close to one another. A new data point is classified using the K-NN algorithm based on similarity after all the existing data has been stored. This means that utilising the K-NN method, fresh data may be quickly and accurately sorted into a suitable category (What is the K-nearest neighbors algorithm?).

K-nearest neighbours classification is straightforward to comprehend and use, in contrast to classification using artificial neural networks. When the data points are well-defined or non-linear, it works well.

In order to decide the class of an unobserved observation, KNN essentially uses a voting method. This indicates that the class that receives the most votes will be the class for the relevant data point.

If K is equal to 1, we will only consider a data point's closest neighbour when determining its class. The ten closest neighbours will be used if K is equal to ten, and so on (Joby).

1. Previous Implementation

P. Sompoch identified traffic signs using the Euclidean distance between samples and reference templates while using the closest neighbour approach. Principal component analysis was used by H. Fleyeh and E. Davami to extract the most useful feature vector from the data. They then generated a set of weights and used Euclidean distance to categorise unknown samples.
The connection between vectors cannot be eliminated by distance measurement, it increases the likelihood of accurate classification outcomes(Fu & Huang, 2010).

2. Advantages

It's straightforward to apply and comprehend.
Both classification and regression issues may be solved with it.
Since there are no underlying assumptions, it is perfect for non-linear data.
Multi-class scenarios may be handled naturally by it.
When given sufficient representative data, it can perform well (Joby).

3. Disadvantages

When the training set is huge, KNN's run-time performance might be subpar. Because all aspects contribute to similarity and subsequently to classification, it is extremely sensitive to irrelevant or duplicate features. This can be prevented by thoughtful feature selection or feature weighting. This approach also has the following two drawbacks: (1) It is unclear which type of distance and which attribute will yield the best results; (2) The computation cost is fairly expensive since we must compute the distance between each query instance and each training sample (Application of K-nearest neighbor (KNN) approach for predicting ...).

4. Input data required and expected output

KNN requires a collection of labelled feature vectors as input data, which may be retrieved from picture data using methods including texture, shape, and colour. A categorization into one of the specified traffic sign classes is the desired outcome. The performance of the technique, which is straightforward to construct, significantly depends on the calibre of the input feature vectors, the value of k, the selection of the distance metric, and the variety and size of the training and testing datasets.

Evaluating the KNN model's performance can be done by using metrics such as accuracy, precision, recall, F1-score, which are calculated using the confusion matrix. The value of k can be chosen by using techniques such as cross-validation to find the best value that gives the highest accuracy.

Although KNN is thought to be computationally cheap, query time grows as the size of the training set does. In conclusion, KNN can be useful for recognising traffic signs, but its effectiveness depends heavily on the quality of the input characteristics and the chosen parameters

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from PIL import Image
import os
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from keras.models import Sequential, load_model
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout

data = []
labels = []
classes = 43
cur_path = os.getcwd()
```

```python
#Retrieving the images and their labels
for i in range(classes):
    path = os.path.join(cur_path,'archive/Train',str(i))
    images = os.listdir(path)

    for a in images:
        try:
            image = Image.open(path + '/'+ a)
            image = image.resize((30,30))
            image = np.array(image)
            #sim = Image.fromarray(image)
            data.append(image)
            labels.append(i)
        except:
            print("Error loading image")
```

```python
#label of each available class
label = {
    0:                               "Speed limit (20km/h)",
    1:                               "Speed limit (30km/h)",
    2:                               "Speed limit (50km/h)",
    3:                               "Speed limit (60km/h)",
    4:                               "Speed limit (70km/h)",     
    5:                               "Speed limit (80km/h)",     
    6:                        "End of speed limit (80km/h)",
    7:                              "Speed limit (100km/h)",
    8:                              "Speed limit (120km/h)",
    9:                                         "No passing",
    10:      "No passing for vechiles over 3.5 metric tons",
    11:             "Right-of-way at the next intersection",
    12:                                     "Priority road",
    13:                                             "Yield",
    14:                                              "Stop",
    15:                                       "No vechiles",
    16:          "Vechiles over 3.5 metric tons prohibited",
    17:                                          "No entry",
    18:                                   "General caution",
    19:                       "Dangerous curve to the left",
    20:                      "Dangerous curve to the right",
    21:                                      "Double curve",
    22:                                        "Bumpy road",
    23:                                     "Slippery road",
    24:                         "Road narrows on the right",
    25:                                         "Road work",
    26:                                   "Traffic signals",
    27:                                       "Pedestrians",
    28:                                 "Children crossing",
    29:                                 "Bicycles crossing",
    30:                                "Beware of ice/snow",
    31:                             "Wild animals crossing",
    32:               "End of all speed and passing limits",
    33:                                  "Turn right ahead",
    34:                                   "Turn left ahead",
    35:                                        "Ahead only",
    36:                              "Go straight or right",
    37:                               "Go straight or left",
    38:                                        "Keep right",
    39:                                         "Keep left",
    40:                              "Roundabout mandatory",
    41:                                 "End of no passing",
    42:"End of no passing by vechiles over 3.5 metric tons"
}
```

```python
#Converting lists into numpy arrays
data = np.array(data)
labels = np.array(labels)

print(data.shape, labels.shape)
#Splitting training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

#Converting the labels into one hot encoding
y_train = to_categorical(y_train, 43)
y_test = to_categorical(y_test, 43)
```

```python
y_test
```

```python
#Building the model
model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=X_train.shape[1:]))
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Dropout(rate=0.25))
model.add(Flatten())
#hidden layer
model.add(Dense(256, activation='relu'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax'))
```

```python
#Compilation of the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

epochs = 15
history = model.fit(X_train, y_train, batch_size=32, epochs=epochs, validation_data=(X_test, y_test))
model.save("traffic_sign_model.h5")


```

```python
#plotting graphs for accuracy
plt.figure(0)
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
plt.figure(1)
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

## Confusion Matrix
In the traffic sign detection system, a confusion matrix is a table that is used to define the performance of the classifier. The confusion matrix is typically used to describe the performance of a classification algorithm by comparing the predicted class labels with the true class labels.
Each row of the matrix represents the instances in a predicted class while each column represents the instances in an actual class (or vice versa). The confusion matrix will have one column and one row for each class in the dataset.
A good classifier will have the majority of the predictions on the diagonal of the matrix, meaning that the predicted class and the true class match. while off-diagonal entries indicate that the classifier made incorrect predictions.
The confusion matrix below shows that majority of the predictions are on the diagonal thus proving that this is a good classifier

```python
#confusion matrix
cmat = confusion_matrix(classes_x, labels)
plt.figure(figsize=(16,16))
sns.heatmap(cmat, annot=True, cbar=False, fmt="d", xticklabels=list(label.values()), yticklabels=list(label.values()))
plt.show()
```

## Classification Report
A classification report is a document used to assess the effectiveness of a classification model in artificial intelligence. Usually, it contains the following metrics:
1. Precision: The proportion of true positive predictions among all positive predictions (true positives plus false positives).
2. Recall: The proportion of true positive predictions among all actual positive observations (true positives plus false negatives).
3. F1-score: The harmonic mean of precision and recall.
4. Support: The number of observations for each class.

The categorization report is frequently displayed as a table, with one row for each class and columns for the metrics.
The below classification report states that the average for precision,recall,f1 score and support is over 90 concluding that the expected results could be achieved on most occassions.

```python
#classification report
print(classification_report(labels, classes_x, target_names = list(label.values())))
```
